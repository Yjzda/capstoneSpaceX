This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

Create a decision tree classifier object then  create a  <code>GridSearchCV</code> object  <code>tree_cv</code> with cv = 10.  Fit the object to find the best parameters from the dictionary <code>parameters</code>.

## TASK  12


Find the method performs best:


```python
# Assuming you have already trained GridSearchCV objects for different models

# Create a dictionary of model names and their respective GridSearchCV objects
model_dict = {
    'Logistic Regression': lr,
    #'Support Vector Machine': svm,
    'knn':KNN,
    'Decision Tree':tree
   
}

# Initialize variables to keep track of the best model and its score
best_model = None
best_score = -1  # Initialize with a low value

# Loop through the models and find the best one based on mean cross-validated score
for model_name, model_cv in model_dict.items():
    mean_score = model_cv.cv_results_['mean_test_score'].max()
    if mean_score > best_score:
        best_score = mean_score
        best_model = model_name

# Print the best model and its cross-validated score
print("Best model:", best_model)
print("Best mean cross-validated score:", best_score)
```
